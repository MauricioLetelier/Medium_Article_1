```python
from cryptocompy import coin
import pandas as pd

coin_data = coin.get_coin_snapshot("ETH", "TUSD")
df = pd.DataFrame(coin_data["Exchanges"])
df["VOLUME24HOUR"] = pd.to_numeric(df["VOLUME24HOUR"])
df["VOLUME24HOURTO"] = pd.to_numeric(df["VOLUME24HOURTO"])
df["MEANPRICE"] = df["VOLUME24HOURTO"] / df["VOLUME24HOUR"]
df = df[
    [
        "MARKET",
        "LOW24HOUR",
        "HIGH24HOUR",
        "PRICE",
        "VOLUME24HOUR",
        "VOLUME24HOURTO",
        "MEANPRICE",
    ]
]
df.sort_values(by="VOLUME24HOUR", ascending=False).head()
```
